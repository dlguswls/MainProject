```python
# 텍스트 전처리 함수
# 2. 특수문자 제거 및 대문자를 소문자로 변경, 
# 4. 문장을 단어 단위로 구분
# 5. 불용어 제거

import re
import nltk
import pickle
from tqdm import tqdm 
#nltk.download('all')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#df['clean_text'] = pd.Series(dtype='object')
df['stopwords_after'] = pd.Series(dtype='object')
#df['pos_tag'] = pd.Series(dtype='object')
#df['NN'] = pd.Series(dtype='object')
df['lemmatization'] = pd.Series(dtype='object')

stops = set(stopwords.words('english'))

def clean_text(texts): 
    corpus = []
    
    for i in tqdm(range(0, len(texts))):
        
        review = texts[i]
        
        review = re.sub('[^a-zA-Z]', ' ', review) # 특수문자 제거 
        review = review.lower() # 대문자를 소문자로 변경, 문장을 단어 단위로 구분
        review = word_tokenize(review)
        
        no_stops = [word for word in review if not word in stops] # 불용어 제거
        df['stopwords_after'][i] = no_stops

    return no_stops
```

```python
clean_text(df['patent_abstract'])
```

```python
df['pos_tag'] = pd.Series(dtype='object')
df['pos_tag'] = df['stopwords_after'].apply(lambda x : nltk.pos_tag(x))
df
```

```python
from tqdm import tqdm, tqdm_pandas

def ext_nn(tokens_pos):
  NN_words = [] # 명사는 NN을 포함하고 있음을 알 수 있음(명사만 추출)
  for word, pos in tokens_pos:
    if 'NN' in pos:
      NN_words.append(word)
  return NN_words

tqdm_pandas(tqdm())

df['NN'] = pd.Series(dtype='object')
df['NN'] = df['pos_tag'].progress_apply(lambda x : ext_nn(x))
df
```

```python
def lemm(NN_words):
  wlem = nltk.WordNetLemmatizer() # Lemmatization(원형(lemma) 찾기) # nltk에서 제공되는 WordNetLemmatizer을 이용
  lemmatized_words = []
  for word in NN_words:
    new_word = wlem.lemmatize(word)
    lemmatized_words.append(new_word)
  return lemmatized_words

tqdm_pandas(tqdm())

df['lemmatization'] = pd.Series(dtype='object')
df['lemmatization'] = df['NN'].progress_apply(lambda x : lemm(x))
df
```

```python
df = df[['patent_number', 'patent_abstract', 'stopwords_after','pos_tag', 'NN', 'lemmatization']]
```

```python
## i, x 등 길이가 2 이하인 문자 제거
from tqdm import tqdm

for i in tqdm(range(0,len(df['lemmatization']))):
    for j in df['lemmatization'][i]:
        if (len(j) < 3):
            df['lemmatization'][i].remove(j)
```

```python
# 빈 리스트를 가지는 행 삭제
empty_idx = df[df['lemmatization'].apply(lambda x: len(x) == 0) == True].index

df.drop(empty_idx, axis=0, inplace=True)
df = df.reset_index()
df.drop('index',axis=1,inplace=True)
df
```
