## 1) 잘 분류 안된 특허

```python
import pandas as pd

df = pd.read_csv('/content/gdrive/MyDrive/Capstone/data/xai_results.csv')
pred = pd.read_csv('/content/gdrive/MyDrive/Capstone/data/wd(0.005)_lr(1e-5)_raw_epoch6_pred_label.csv', index_col = 0)
true = pd.read_csv('/content/gdrive/MyDrive/Capstone/data/xai_data.csv', index_col = 0)
pred = pred.iloc[true['index'].values.tolist(),:].reset_index()
df
```

## 1-1) 1인데 0으로 예측

```python
from tqdm import tqdm
cols = true.columns
find_list = []
for k in tqdm(range(len(true))) : 
  tr_li = true.iloc[k,2:].values.tolist()
  pre_li = pred.iloc[k,2:].values.tolist()
  error_colnames = []
  for i in range(len(tr_li)) : 
    if tr_li[i] == 1 : 
      if pre_li[i] == 0 : 
        error_colnames.append(cols[i+2])
  find_list.append(error_colnames)
```

```python
from ast import literal_eval
wordcloud_word = {}

for f in tqdm(range(len(true))) : 
  vals = df.loc[f,find_list[f]].values
  indx = df.loc[f,find_list[f]].index
  try : 
    val_d = pd.DataFrame()

    words = []
    aa = literal_eval(vals[0])
    for i in range(len(aa)) : 
      words.append(aa[i][0])
    val_d.index = words

    for i in range(len(vals)) : 
      val = literal_eval(vals[i])
      ind = indx[i]
      nums = []
      for j in val : 
        nums.append(j[1])
      val_d[ind] = nums

    for j in val_d.columns : 
      words_lists =  val_d.sort_values(j, ascending = False)[:5].index.tolist()
      for i in range(len(words_lists)) : 
        if wordcloud_word.get(words_lists[i]) == None : 
          wordcloud_word[words_lists[i]] = 1
        else : 
          wordcloud_word[words_lists[i]] += 1
  except : 
    pass
```
