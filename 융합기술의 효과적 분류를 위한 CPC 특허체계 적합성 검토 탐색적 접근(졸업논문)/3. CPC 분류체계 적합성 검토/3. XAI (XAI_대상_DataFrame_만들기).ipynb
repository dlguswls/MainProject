```python
import pandas as pd
data = pd.read_csv("/content/gdrive/MyDrive/Project/Capstone/data/wd(0.005)_lr(1e-5)_raw_epoch6_pred_prob_classification.csv",index_col=0)

true = pd.read_csv("/content/gdrive/MyDrive/Project/Capstone/data/test_label.csv",index_col=0) 
true.reset_index(inplace=True)
true.drop('index',axis=1,inplace=True)

pred = pd.read_csv("/content/gdrive/MyDrive/Project/Capstone/data/wd(0.005)_lr(1e-5)_raw_epoch6_pred_label.csv",index_col=0) 
pred.reset_index(inplace=True)
pred.drop('index',axis=1,inplace=True)
```

```python
df = data[data['classification']==0]
df = true.loc[df.index,:]
df
```

```python
true.iloc[:,1:].sum(axis=0).sort_values(ascending=False).sum()
```

```python
# test셋 전체 특허들 중 true label별 개수 확인
probas = pd.Series(true.iloc[:,1:].sum(axis=0).sort_values(ascending=False)/35161)
```

```python
probas
```

```python
probas['H01B']
```

```python
labels =true.iloc[:,1:].columns.tolist()
len(labels)
```

```python
from tqdm import tqdm
index_list = []
for i in tqdm(labels) : 
  try:
    a = df[df[i]==1]
    aa = a.sample(n = int(1000*probas[i])).index.tolist()
    for j in aa : 
      index_list.append(j)
  except : 
    pass
```

```python
index_list = sorted(list(set(index_list)))
```

```python
final = df.loc[index_list,:].reset_index()
```
