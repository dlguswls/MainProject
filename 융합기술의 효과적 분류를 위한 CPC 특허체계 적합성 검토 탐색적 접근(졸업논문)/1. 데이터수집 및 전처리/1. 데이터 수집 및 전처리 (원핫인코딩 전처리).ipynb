## label one-hot encoding

```python
subclass = []
for i in range(len(df)) : 
  subclass_lists = df['cpc_group_id'][i].split(',')
  for j in range(len(subclass_lists)) : 
    if subclass_lists[j][0]=='H' : 
      if subclass_lists[j] in subclass : 
        pass
      else : 
        subclass.append(subclass_lists[j])

subclass = sorted(subclass)
```

```python
cols = ['patent_number', 'patent_abstract']
cols = cols + subclass
cols

data = pd.DataFrame(columns = cols)
data['patent_number'] = df['patent_number']
data['patent_abstract'] = df['patent_abstract']
data = data.fillna(0)
```

```python
for i in tqdm(range(len(df))) : 
  subclass_lists = df.loc[i, 'cpc_group_id'].split(',')
  subclass_lists = set(subclass_lists) 
  subclass_lists = list(subclass_lists) 
  for k in subclass_lists : 
    if k[0] =='H' : 
      data.loc[i, k] = 1
    else :
      pass
```

```python
data.to_csv('/content/gdrive/MyDrive/Project/Capstone/data/H_section_onehot.csv')
```
