## 3) 분류 안된 데이터셋 중 몇개 틀린지 확인해보기

```python
true_X = true.iloc[class_X.index,1:]
true_X
```

```python
from tqdm import tqdm

true_X['class_list'] = pd.Series(dtype='object')

for i in tqdm(true_X.index):
  class_list = []
  for j in true_X.columns[:-1]:
      class_list.append(true_X.loc[i][j])
  true_X['class_list'][i] = class_list
```

```python
true_X
```

```python
pred_X = pred.iloc[class_X.index,1:]
pred_X
```

```python
from tqdm import tqdm

pred_X['class_list_pred'] = pd.Series(dtype='object')

for i in tqdm(pred_X.index):
  class_list = []
  for j in pred_X.columns[:-1]:
      class_list.append(pred_X.loc[i][j])
  pred_X['class_list_pred'][i] = class_list
```

```python
df = pd.concat([true_X,pred_X[['class_list_pred']]],axis=1)
df
```

```python
from tqdm import tqdm

df['X_num'] = pd.Series(dtype='int')

for i in tqdm(df.index):
  num = 0
  for j in range(0,52):
    if (df.loc[i]['class_list'][j] == df.loc[i]['class_list_pred'][j]):
      pass
    else:
      num += 1
  df['X_num'][i] = num
```
