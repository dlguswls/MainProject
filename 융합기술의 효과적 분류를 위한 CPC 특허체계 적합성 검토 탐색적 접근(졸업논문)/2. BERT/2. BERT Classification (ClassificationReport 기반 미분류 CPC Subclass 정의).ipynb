```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = pd.read_csv('/content/drive/MyDrive/Capstone/data/wd(0.005)_lr(1e-5)_raw_epoch6_classification_report_0.csv')
df1 = pd.read_csv('/content/drive/MyDrive/Capstone/data/wd(0.005)_lr(1e-5)_raw_epoch6_classification_report_1.csv')
```

```python
df0 = df0.sort_values('support')
df1 = df1.sort_values('support')
```

```python
plt.figure(figsize = (20, 8))
plt.plot(df0['support'].values.tolist(),df0['precision'].values.tolist(), label = 'precision')
plt.plot(df0['support'].values.tolist(),df0['recall'].values.tolist(),label = 'recall')
plt.plot(df0['support'].values.tolist(),df0['f1-score'].values.tolist(),label = 'f1-score')
plt.legend(fontsize = 15)
```

```python
plt.figure(figsize = (20, 8))
plt.plot(df0['Unnamed: 0'].values.tolist(),df0['precision'].values.tolist(), label = 'precision')
plt.plot(df0['Unnamed: 0'].values.tolist(),df0['recall'].values.tolist(),label = 'recall')
plt.plot(df0['Unnamed: 0'].values.tolist(),df0['f1-score'].values.tolist(),label = 'f1-score')
plt.legend(fontsize = 15)
```
