## 공원 면적

```python
plt.figure(figsize=(20,10))
fig, axs = plt.subplots(1,2)
sns.boxplot(gs_area['면적'], ax = axs[0]).set_title('강서구 공원 면적 분포')
sns.boxplot(sp_area['면적'], ax = axs[1]).set_title('송파구 공원 면적 분포')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(gs_area['면적'])
plt.title('강서구 공원 면적 분포')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(sp_area['면적'])
plt.title('송파구 공원 면적 분포')
plt.show()
```
