## 공원 면적

```python
gs_area = pd.read_csv('C:/Users/lkkcl/ewha/공공빅데이터_인턴/4. 실무형 프로젝트/서울2 6조/data/입지선정/면적/강서구경도.csv', encoding='cp949')
sp_area = pd.read_csv('C:/Users/lkkcl/ewha/공공빅데이터_인턴/4. 실무형 프로젝트/서울2 6조/data/입지선정/면적/송파구경도.csv', encoding='cp949')
```

```python
gs_area.describe()
```

```python
pd.options.display.float_format = '{:.5f}'.format
sp_area.describe()
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(gs_area['면적'])
plt.title('강서구 공원 면적 분포')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(sp_area['면적'])
plt.title('송파구 공원 면적 분포')
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,7))
sns.boxplot(gs_area['면적'],ax=ax[0])
plt.title('강서구 공원 면적 분포')

sns.boxplot(sp_area['면적'],ax=ax[1])
plt.title('송파구 공원 면적 분포')

plt.show()
```

```python
# 송파구 공원 면적 700m2 이하 제거
sp_area = sp_area[sp_area['면적'] > 700]

# 강서구 공원 면적 700m2 이하 제거
gs_area = gs_area[gs_area['면적'] > 700]

# 저장
sp_area.to_csv('C:/Users/lkkcl/ewha/공공빅데이터_인턴/4. 실무형 프로젝트/서울2 6조/data/입지선정/면적/송파구면적_700이상.csv', encoding='euc-kr')
gs_area.to_csv('C:/Users/lkkcl/ewha/공공빅데이터_인턴/4. 실무형 프로젝트/서울2 6조/data/입지선정/면적/강서구면적_700이상.csv', encoding='euc-kr')
```
