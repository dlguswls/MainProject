```python
# import pandas as pd
# import json, openpyxl


def save(df,filename):
    writer = pd.ExcelWriter(filename)
    df.to_excel(writer,"sheet1")
    writer.save()
    
# jsonString = open("./nonPayment.json").read()   
df= pd.read_json("./nonPayment.json")
print(df.count())   

start = bgnde
end = endde

save(df,f"animal_2_{start}_{end}.xlsx")
```
